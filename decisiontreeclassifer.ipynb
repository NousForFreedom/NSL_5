```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

x= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/y0uC4N/preprocessing/updated_type10.csv')
dt_clf = DecisionTreeClassifier(max_depth = 5,random_state=156)
parameters = {'max_depth':[1,2,3,4,5],'min_samples_split':[2,3]}
grid_dtree = GridSearchCV(dt_clf,param_grid=parameters,cv=3,refit=True)
x.head()
```

```python
y=x['type']
x=x.drop('type',axis=1)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=121)
x_train = x_train.drop('isCan0',axis=1)
x_train = x_train.drop('isService',axis=1)
print(x_train.corr(method = 'pearson'))
x_test = x_test.drop('isCan0',axis=1)
x_test = x_test.drop('isService',axis=1)
```

```python
grid_dtree.fit(x_train,y_train)
scores_df=pd.DataFrame(grid_dtree.cv_results_)

scores_df[['params', 'mean_test_score', 'rank_test_score', \
           'split0_test_score', 'split1_test_score', 'split2_test_score']]

print('GridSearchCV 최적 파라미터:', grid_dtree.best_params_)
print('GridSearchCV 최고 정확도: {0:.4f}'.format(grid_dtree.best_score_))

# GridSearchCV의 refit으로 이미 학습이 된 estimator 반환
estimator = grid_dtree.best_estimator_

# GridSearchCV의 best_estimator_는 이미 최적 하이퍼 파라미터로 학습이 됨
pred = estimator.predict(x_test)
```
